```python
X = df.iloc[:,[2,3]].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss = []
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('No. of Cluster')
plt.ylabel('WCSS Values')
plt.show
```

```python
kmeansmodel = KMeans(n_clusters=5, init='k-means++', random_state = 0)
```

```python
ymeansmodel = kmeansmodel.fit_predict(X)
```

```python
plt.scatter(X[ymeansmodel == 0,0], X[ymeansmodel == 0,1], s=80, c="red", label='Customer 1')
plt.scatter(X[ymeansmodel == 1,0], X[ymeansmodel == 1,1], s=80, c="yellow", label='Customer 2')
plt.scatter(X[ymeansmodel == 2,0], X[ymeansmodel == 2,1], s=80, c="cyan", label='Customer 3')
plt.scatter(X[ymeansmodel == 3,0], X[ymeansmodel == 3,1], s=80, c="black", label='Customer 4')
plt.scatter(X[ymeansmodel == 4,0], X[ymeansmodel == 4,1], s=80, c="green", label='Customer 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=80, c="blue", label='Centroids')
plt.title('Cluster of Customers')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Scores')
plt.legend()
plt.show
```
